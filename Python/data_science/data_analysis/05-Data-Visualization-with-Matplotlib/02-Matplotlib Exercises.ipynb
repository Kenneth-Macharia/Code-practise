```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# For non-jupyter notebook display use: plt.show() after every plot declaration.
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,'b', alpha=0.7)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig_3 = plt.figure()

ax_3 = fig_3.add_axes([0,0,1,1])
ax_4 = fig_3.add_axes([0.2,0.5,.4,.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Large plot
ax_3.plot(x,z,'b-')
ax_3.set_xlabel('X')
ax_3.set_ylabel('Z')

# Insert plot
ax_4.set_xlim([20,22])
ax_4.set_ylim([30,50])
ax_4.plot(x,y,'b-')
ax_4.set_title('zoom')
ax_4.set_xlabel('X')
ax_4.set_ylabel('Y')

fig_3

fig_4 = plt.figure()
ax_5 = fig_4.add_axes([0,0,1,1])
ax_6 = fig_4.add_axes([0.2,0.5,.4,.4])

ax_5.plot(x,z,'b-')
ax_5.set_xlabel('X')
ax_5.set_ylabel('Z')

ax_6.set_xlim([20,22])
ax_6.set_ylim([30,50])
ax_6.plot(x,y,'b-')
ax_6.set_title('zoom')
ax_6.set_xlabel('X')
ax_6.set_ylabel('Y')

fig_4
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fg, ax = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
line, = ax[0].plot(x, y, ls='--', color='blue', lw=3) # with custom dash
line.set_dashes([2, 2, 2, 2])    # format: line length, space length, ...
ax[1].plot(x, z, ls='-', color='red', lw=3)
fg
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fg, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

ax[0].plot(x, y, ls='-', color='blue', lw=5)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

line, = ax[1].plot(x, z, ls='--', color='red', lw=3)
line.set_dashes([2, 2, 2, 2])
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
```
